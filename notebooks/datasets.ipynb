```
from datasets import load_dataset
from rich.table import Table
from rich.console import Console
import tiktoken
from tqdm import tqdm
```

```
# Load the entire dataset
dataset = load_dataset("ai-hyz/MemoryAgentBench")
```

```
console = Console()
```

# Accurate Retrieval

This dataset focuses on evaluating the retrieval capabilities of memory agents. It contains tasks that require agents to retrieve relevant information accurately from a given context.

```
ar = dataset["Accurate_Retrieval"]
```

```
unique_sources = set([row["metadata"]["source"] for row in ar])
```

```
unique_sources
```

```
encoder = tiktoken.get_encoding("o200k_base")

results = []

for sample in tqdm(ar):
    context_text = sample["context"]
    source_name = sample["metadata"]["source"]
    context_token_count = len(encoder.encode(context_text))
    results.append(
        {
            "source": source_name,
            "questions": len(sample["questions"]),
            "answers": len(sample["answers"]),
            "context_chars": len(context_text),
            "context_tokens": context_token_count,
        }
    )
```

```
table = Table("Source", "Questions", "Answers", "Chars", "Tokens", show_lines=True)

for row in results:
    table.add_row(
        row["source"],
        str(row["questions"]),
        str(row["answers"]),
        str(row["context_chars"]),
        f"{round(row['context_tokens'] / 1000)} K tokens",
    )

console.print(table)
```
