```
from tape_mem.agents.tape import TapeAgent
from tape_mem.types.agent import ConversationMessage, ConversationSession
from tape_mem_cli.settings.env import Env
from loguru import logger
```

```
env = Env()  # ty:ignore[missing-argument]
```

```
logger.info(f"using llm endpoint: {env.openai_compatible_base_url}")
logger.info(f"using llm model: {env.llm_model}")
```

```
from tape_mem.types.provider import ProviderConfig

provider = ProviderConfig(
    model=env.llm_model,
    base_url=env.openai_compatible_base_url,
    api_key=env.openai_compatible_api_key,
)
```

```
from typing import Literal, Sequence
from datetime import datetime
from dataclasses import dataclass, field


@dataclass(slots=True)
class Message(ConversationMessage):
    """Concrete implementation of ConversationMessage."""

    role: Literal["user", "assistant", "system"]
    content: str


@dataclass(slots=True)
class Session(ConversationSession):
    """Concrete implementation of ConversationSession."""

    chat_time: datetime
    messages: Sequence[Message] = field(default_factory=list)
```

```
from tape_mem.dataset.templates import LongMemEvalTemplate

template = LongMemEvalTemplate()
agent = TapeAgent(provider=provider, template=template)
```

```
sessions = [
    Session(
        chat_time=datetime.now(),
        messages=[
            Message(role="user", content="What is the capital of France?"),
            Message(role="assistant", content="The capital of France is Paris."),
        ],
    )
]
```

```
agent.memorize_conversation(
    sessions=sessions,
)
```
